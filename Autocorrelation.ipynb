### LABEL

```python
replace = {"Background": 1, "Botnet":2}

selected_df['Label'] = selected_df['Label'].apply(lambda x: str(x))

for value in selected_df['Label']:
    if "Background" in value:
        value = 2
    elif "Botnet" in value:
        value = 1
    else:
        value = 3

selected_df.head(10)
```

### PROTO

```python
Proto_all_types = selected_df['Proto'].unique()
print(Proto_all_types)
```

```python
replace = {}

print(selected_df['Proto'].dtype)

for value in selected_df['Proto']:
    if str(value) not in replace.keys():
        replace[str(value)] = 1
    elif str(value) in replace.keys():
        replace[str(value)] += 1

num = 0
replace_new = {}
for x in range(len(replace)):
    max_key = max(replace, key=replace.get)
    num += 1
    replace_new[max_key] = num
    del replace[max_key]

for value in selected_df['Proto']:
    if str(value) in replace_new:
        value = replace_new[str(value)]

selected_df.head(5)

```
